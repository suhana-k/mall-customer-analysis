```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Mall_Customers (1).csv")
print(data.head())
```

```python
print(data.shape)
```

```python
# Count gender
gender_count = data['Gender'].value_counts()

# Plot bar chart
plt.bar(gender_count.index, gender_count.values,color=['yellow', 'pink'])
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")

plt.show()
```

```python
plt.hist(data['Age'], bins=10,color="red")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")

plt.show()
```

```python
plt.hist(data['Annual Income (k$)'], bins=10,color="orange")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Frequency")
plt.title("Income Distribution")

plt.show()
```

```python
# Group spending score into ranges
bins = [0, 20, 40, 60, 80, 100]
labels = ["0-20", "20-40", "40-60", "60-80", "80-100"]

data['Spending Group'] = pd.cut(data['Spending Score (1-100)'], bins=bins, labels=labels)

spending_count = data['Spending Group'].value_counts().sort_index()

plt.bar(spending_count.index, spending_count.values,color="green")
plt.xlabel("Spending Score Group")
plt.ylabel("Count")
plt.title("Spending Score Distribution")

plt.show()
```

```python
#scatter plot income vs spending score

plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("Income vs Spending Score")

plt.show()
```

```python
#outliers detection
data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].boxplot()
plt.show()
```

```python
#genderwise analysis
import pandas as pd

gender_avg = data.groupby('Gender')['Spending Score (1-100)'].mean()

gender_avg.plot(kind='bar', color=['blue','pink'])
plt.ylabel("Average Spending Score")
plt.title("Spending Score by Gender")

plt.show()
```

```python
#age vs spending score
plt.scatter(data['Age'], data['Spending Score (1-100)'])
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Age vs Spending Score")

plt.show()
```

```python
#correlation heatmap
import seaborn as sns

corr = data.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")

plt.show()
```

```python
#pairplot
import seaborn as sns

sns.pairplot(data)
plt.show()
```
